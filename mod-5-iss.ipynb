```python
import requests as req
import matplotlib.pyplot as plt
import pandas as pd
import json
import time

data = []

#range 0 to 100
for i in range(0,100):
    #link
    response = req.get ("http://api.open-notify.org/iss-now.json")
    text = response.text
    obj = json.loads(text)
    
    #for the data
    data.append((obj["timestamp"], obj['iss_position']['longitude'], obj['iss_position']['latitude']))
    #time = 10
    time.sleep(10)
#chart
df = pd.DataFrame(data, columns = ['Timestamp', 'Longitude', 'Latitude'])
print (pd.DataFrame(df))
```

```python
#the picture??? basta the graph thingy
df = pd.DataFrame(data, columns = ['Timestamp', 'Longitude', 'Latitude'])
#latitude down #longitude left #color of dots red #size square
axl = df.plot.scatter(x='Latitude',
                      y='Longitude',
                      c='Red',
                      figsize=(30,30)
                     )
axl.set_title("International Space Station Location")
plt.show()
```
